```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def display_frame(video_path, frame_no):
    # 動画ファイルを開く
    cap = cv2.VideoCapture(video_path)

    # 動画の総フレーム数を取得
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # 指定フレームが存在するかチェック
    if frame_no > frame_count:
        print("Error: frame does not exist")
        return

    # フレーム番号を指定
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_no)

    # フレームを読む
    ret, frame = cap.read()

    if ret:
        # OpenCVは色をBGRで扱うが、MatplotlibはRGBで色を扱うため変換する
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        plt.figure(figsize = (16,9))
        
        # フレームを表示
        plt.imshow(frame)
        plt.axis('on')  # 軸の表示をオンにする
        plt.xticks(range(0, frame.shape[1], 100))  # x軸の目盛り
        
        # y軸の目盛りを反転させる
        plt.yticks(range(0, frame.shape[0], 100))  # y軸の目盛り
        plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(int(frame.shape[0] - x))))
        
        plt.show()
        
    else:
        print("Error: failed to read frame")

    # メモリの解放
    cap.release()

```

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as patches

def display_frame(video_path, frame_no, rectangle_coordinates, circle_center_radius=(960, 540, 125)):
    # 動画ファイルを開く
    cap = cv2.VideoCapture(video_path)

    # 動画の総フレーム数を取得
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # 指定フレームが存在するかチェック
    if frame_no > frame_count:
        print("Error: frame does not exist")
        return

    # フレーム番号を指定
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_no)

    # フレームを読む
    ret, frame = cap.read()

    if ret:
        # OpenCVは色をBGRで扱うが、MatplotlibはRGBで色を扱うため変換する
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # フレームを表示
        fig, ax = plt.subplots(figsize=(16, 9))  # 大きさを指定
        ax.imshow(frame)
        ax.axis('on')  # 軸の表示をオンにする
        ax.set_xticks(range(0, frame.shape[1], 100))  # x軸の目盛り
        
        # y軸の目盛りを反転させる
        ax.set_yticks(range(0, frame.shape[0], 100))  # y軸の目盛り
        ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(int(frame.shape[0] - x))))
        
        # 矩形が指定されている場合、それを描画
        if rectangle_coordinates is not None:
            for rectangle_coordinate in  rectangle_coordinates:
                x1, y1, x2, y2,_ = rectangle_coordinate
                rect = patches.Rectangle((x1, frame.shape[0]-y1), x2-x1, y1-y2, linewidth=1, edgecolor='r', facecolor='none')
                ax.add_patch(rect)
        
       
        # 円が指定されている場合、それを描画
        if circle_center_radius is not None:
            cx, cy, radius = circle_center_radius
            circle = patches.Circle((cx, frame.shape[0]-cy), radius, fill=False, edgecolor='r')
            ax.add_patch(circle)
            
        plt.show()
    else:
        print("Error: failed to read frame")

    # メモリの解放
    cap.release()
```
